```python
# 把特征个数少的列全部丢弃
def drop_fea(df):
    features=df.columns
    feature_selected=[] 
    sz=df.shape[0]
    for feat in features:
        sz_null=np.sum(df[feat].isnull()) #
        ratio=float(sz_null)/sz
        if ratio < 0.98:
            feature_selected.append((feat))  
    return df[feature_selected]
```

```python
def dumyuserfeature(train):
    pre_feat = 'UserInfo_'
    # 检查离散变量和连续变量
    asd = 0
    index=[]
    train_copy = train.copy()
    for i ,col in enumerate(train.columns):
        print(i)
        cofe = len(train.groupby(col).count())
        if cofe < 20:
            if cofe == 2:
                continue
            feikong = np.sum([train[col] != -999] )
            if feikong < len(train) * 0.1:
                continue
                
#             train_copy[col+'_'] =  pd.get_dummies(train[col]
            
            train_copy = train_copy.join(pd.get_dummies(train[col], prefix=col+'_'))
            
            index.append(col)
            asd += 1
#         train_copy.drop(col,inplace=True)
    print(asd,'个离散化的特征')
    return train_copy
```

```python
def dumyuserfeature_2(train,train_1):
    pre_feat = 'UserInfo_'
    # 检查离散变量和连续变量
    asd = 0
    index=[]
    train_copy = train.copy()
    train_1_copy = train_1.copy()
    for i ,col in enumerate(train.columns):
        print(i)
        cofe = len(train.groupby(col).count())
        if cofe < 20:
            if cofe == 2:
                continue
            feikong = np.sum([train[col] != -999] )
            if feikong < len(train) * 0.1:
                continue
                
#             train_copy[col+'_'] =  pd.get_dummies(train[col]
            
            train_copy = train_copy.join(pd.get_dummies(train[col], prefix=col+'_'))
            index.append(col)
            asd += 1
            if col == 'flag':
                continue
            train_1_copy = train_1_copy.join(pd.get_dummies(train_1[col], prefix=col+'_'))
#         train_copy.drop(col,inplace=True)
    print(asd,'个离散化的特征')
    return train_copy,train_1_copy
```

```python
def get_min(a,b):
    c = a.copy()
    for i in range(len(a)):
        if a[i]>b[i]:
            c[i] = b[i]
    return c
```
